```python
from automata import PedestrianConflictAutomataModel, PedestrianConflictParameters
from ipywidgets import interact, IntSlider
from view import build_create_video, plot_system_state
```

# Create simulation

```python
pedestrian_conflict_parameters = PedestrianConflictParameters()
pedestrian_conflict_automata_model = PedestrianConflictAutomataModel(
    pedestrian_conflict_parameters
)
```

```python
steps = pedestrian_conflict_automata_model.run(3)
```

```python
import numpy as np

for row in np.random.random((3, 2)):
    print(" ".join(str(cell) for cell in row))
```

# Plot one step

```python
plot_system_state(
    steps[1],
    pedestrian_conflict_parameters.full_grid_width,
    pedestrian_conflict_parameters.full_grid_length,
)
```

# Make video

```python
# it takes a time to update the videos after an slide change
create_video = build_create_video(
    steps,
    pedestrian_conflict_parameters.full_grid_width,
    pedestrian_conflict_parameters.full_grid_length,
    simulation_interval=100,
)
interact(create_video, n=IntSlider(min=3, max=30, step=1, value=10));
```
